```python
import requests
import datetime
import os

# äº¤æ˜“å¯¹
symbol = "DOGE-USDT-SWAP-aggtrades"

# ä¸‹è½½ç›®å½•
download_dir = "DOGE-USDT-SWAP-aggtrades-data"
os.makedirs(download_dir, exist_ok=True)

# è®¡ç®—è¿‡å»ä¸€å¹´çš„æ—¥æœŸ
start_date = datetime.date(2024, 12, 16)
end_date = datetime.date.today()

# HTTP è¯·æ±‚å¤´ï¼Œæ¨¡æ‹Ÿæµè§ˆå™¨è®¿é—®
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# å¾ªç¯ä¸‹è½½æ–‡ä»¶
current_date = start_date
while current_date <= end_date:
    date_str_dir = current_date.strftime("%Y%m%d")  # ç›®å½•æ ¼å¼ 20240220
    date_str_file = current_date.strftime("%Y-%m-%d")  # æ–‡ä»¶æ ¼å¼ 2024-02-20

    # ç”Ÿæˆæ–‡ä»¶è·¯å¾„
    file_name = f"{symbol}-trades-{date_str_file}.zip"
    file_path = os.path.join(download_dir, file_name)

    # ç”Ÿæˆä¸‹è½½ URL
    file_url = f"https://www.okx.com/cdn/okex/traderecords/aggtrades/daily/{date_str_dir}/{symbol}-{date_str_file}.zip"

    print(f"æ­£åœ¨ä¸‹è½½: {file_url}")

    # å‘é€è¯·æ±‚
    response = requests.get(file_url, headers=headers, stream=True)

    if response.status_code == 200:
        with open(file_path, "wb") as f:
            for chunk in response.iter_content(chunk_size=1024):
                f.write(chunk)
        print(f"ä¸‹è½½å®Œæˆ: {file_path}")
    else:
        print(f"æ–‡ä»¶ä¸å­˜åœ¨ ({response.status_code}): {file_url}")

    # é€’å¢æ—¥æœŸ
    current_date += datetime.timedelta(days=1)
```

```python
import os
import pandas as pd
import warnings

# ç¦ç”¨æ‰€æœ‰è­¦å‘Š
warnings.filterwarnings("ignore")

# è®¾ç½® CSV æ–‡ä»¶æ‰€åœ¨ç›®å½•
data_dir = r"C:\Users\qianz\Desktop\Untitled Folder\BTC-USDT-SWAP-aggtrades-data"

# åˆå§‹åŒ–ç»“æœåˆ—è¡¨
results_list = []

# éå†æ‰€æœ‰ CSV æ–‡ä»¶
for file in os.listdir(data_dir):
    if file.endswith(".csv") and "BTC-USDT-SWAP-aggtrades" in file:  # è¿‡æ»¤ç¬¦åˆå‘½åè§„åˆ™çš„æ–‡ä»¶
        file_path = os.path.join(data_dir, file)

        # **ä¿®æ”¹: æå–å®Œæ•´æ—¥æœŸ (YYYY-MM-DD)**
        date_str = file.split("-")[-3:]  # æå–æœ€å 3 æ®µ ['YYYY', 'MM', 'DD.csv']
        date_str = "-".join(date_str).replace(".csv", "")  # å˜æˆ YYYY-MM-DD
        print(f"ğŸ”„ æ­£åœ¨å¤„ç† {file} ... æ—¥æœŸ: {date_str}")

        # é¢„å¤„ç† CSV æ•°æ®ï¼Œè½¬æ¢ä¸º K çº¿æ ¼å¼
        kline_processor = KlineProcessor(file_path)
        kline_df = kline_processor.get_kline_df()

        # è¿è¡Œç­–ç•¥
        strategy = VolatilityStrategy(kline_df)
        strategy_df = strategy.get_strategy_df()

        # è¿è¡Œå›æµ‹
        print("â–¶ï¸ å¼€å§‹å›æµ‹...")
        bt = Backtest(
            df=strategy_df,
            initial_balance=10000,
            leverage=5,
            open_fee=0.0005,
            close_fee=0.0005,
            tp_ratio=0.1,
            sl_ratio=0.05
        )

        # è·å–å›æµ‹ç»“æœ
        results_df = bt.get_results()

        # **æ–°å¢åˆ¤æ–­ï¼šå¦‚æœ results_df ä¸ºç©ºï¼Œåˆ™å½“æ—¥æ”¶ç›Šä¸º 0**
        if results_df.empty:
            print(f"âš ï¸ {date_str}: æ— äº¤æ˜“è®°å½•ï¼Œæ”¶ç›Šä¸º 0")
            final_balance = 10000
            profit = 0
            profit_rate = 0
        else:
            final_balance = results_df["final_balance"].iloc[-1]
            profit = final_balance - 10000  # è®¡ç®—æ”¶ç›Šé‡‘é¢
            profit_rate = profit / 10000  # è®¡ç®—æ”¶ç›Šç‡

        # è®°å½•ç»“æœ
        results_list.append({
            "date": date_str,  # è®°å½•å®Œæ•´æ—¥æœŸ
            "final_balance": final_balance,
            "profit": profit,
            "profit_rate": profit_rate
        })

# è½¬æ¢ä¸º DataFrame
final_results_df = pd.DataFrame(results_list)

# **ç¡®ä¿ `date` åˆ—å­˜å‚¨ä¸º datetime ç±»å‹**
final_results_df["date"] = pd.to_datetime(final_results_df["date"], format="%Y-%m-%d")

# **æŒ‰æ—¥æœŸå‡åºæ’åº**
final_results_df = final_results_df.sort_values(by="date").reset_index(drop=True) 
```
