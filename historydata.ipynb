```python
import requests
import datetime
import os

# 交易对
symbol = "DOGE-USDT-SWAP-aggtrades"

# 下载目录
download_dir = "DOGE-USDT-SWAP-aggtrades-data"
os.makedirs(download_dir, exist_ok=True)

# 计算过去一年的日期
start_date = datetime.date(2024, 12, 16)
end_date = datetime.date.today()

# HTTP 请求头，模拟浏览器访问
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# 循环下载文件
current_date = start_date
while current_date <= end_date:
    date_str_dir = current_date.strftime("%Y%m%d")  # 目录格式 20240220
    date_str_file = current_date.strftime("%Y-%m-%d")  # 文件格式 2024-02-20

    # 生成文件路径
    file_name = f"{symbol}-trades-{date_str_file}.zip"
    file_path = os.path.join(download_dir, file_name)

    # 生成下载 URL
    file_url = f"https://www.okx.com/cdn/okex/traderecords/aggtrades/daily/{date_str_dir}/{symbol}-{date_str_file}.zip"

    print(f"正在下载: {file_url}")

    # 发送请求
    response = requests.get(file_url, headers=headers, stream=True)

    if response.status_code == 200:
        with open(file_path, "wb") as f:
            for chunk in response.iter_content(chunk_size=1024):
                f.write(chunk)
        print(f"下载完成: {file_path}")
    else:
        print(f"文件不存在 ({response.status_code}): {file_url}")

    # 递增日期
    current_date += datetime.timedelta(days=1)
```

```python
import os
import pandas as pd
import warnings

# 禁用所有警告
warnings.filterwarnings("ignore")

# 设置 CSV 文件所在目录
data_dir = r"C:\Users\qianz\Desktop\Untitled Folder\BTC-USDT-SWAP-aggtrades-data"

# 初始化结果列表
results_list = []

# 遍历所有 CSV 文件
for file in os.listdir(data_dir):
    if file.endswith(".csv") and "BTC-USDT-SWAP-aggtrades" in file:  # 过滤符合命名规则的文件
        file_path = os.path.join(data_dir, file)

        # **修改: 提取完整日期 (YYYY-MM-DD)**
        date_str = file.split("-")[-3:]  # 提取最后 3 段 ['YYYY', 'MM', 'DD.csv']
        date_str = "-".join(date_str).replace(".csv", "")  # 变成 YYYY-MM-DD
        print(f"🔄 正在处理 {file} ... 日期: {date_str}")

        # 预处理 CSV 数据，转换为 K 线格式
        kline_processor = KlineProcessor(file_path)
        kline_df = kline_processor.get_kline_df()

        # 运行策略
        strategy = VolatilityStrategy(kline_df)
        strategy_df = strategy.get_strategy_df()

        # 运行回测
        print("▶️ 开始回测...")
        bt = Backtest(
            df=strategy_df,
            initial_balance=10000,
            leverage=5,
            open_fee=0.0005,
            close_fee=0.0005,
            tp_ratio=0.1,
            sl_ratio=0.05
        )

        # 获取回测结果
        results_df = bt.get_results()

        # **新增判断：如果 results_df 为空，则当日收益为 0**
        if results_df.empty:
            print(f"⚠️ {date_str}: 无交易记录，收益为 0")
            final_balance = 10000
            profit = 0
            profit_rate = 0
        else:
            final_balance = results_df["final_balance"].iloc[-1]
            profit = final_balance - 10000  # 计算收益金额
            profit_rate = profit / 10000  # 计算收益率

        # 记录结果
        results_list.append({
            "date": date_str,  # 记录完整日期
            "final_balance": final_balance,
            "profit": profit,
            "profit_rate": profit_rate
        })

# 转换为 DataFrame
final_results_df = pd.DataFrame(results_list)

# **确保 `date` 列存储为 datetime 类型**
final_results_df["date"] = pd.to_datetime(final_results_df["date"], format="%Y-%m-%d")

# **按日期升序排序**
final_results_df = final_results_df.sort_values(by="date").reset_index(drop=True) 
```
